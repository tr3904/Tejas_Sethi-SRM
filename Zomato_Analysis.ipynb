# Objective: Zomato Data Analysis

```python
# Importing libraries
# pandas: Imported for data manipulation and analysis
# numpy: Imported for data operations
# matplotlib, seaborn, and plotly: Imported for data visualization
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv("Zomato data .csv")
df1
```

```python
# Data Cleaning
df1.isnull().sum()
```

```python
df1
```

```python
def operate(values):
    values = str(values).split("/")
    sat = values[0]
    return float(sat)
df1["rate"] = df1["rate"].apply(operate)
df1.head(10)
```

```python
df1
```

```python
df1.select_dtypes(include=["object"])
```

# What Type of restaurants do the majority of customers order from?

```python
df1
```

```python
# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df1,x=df1["listed_in(type)"],color="green")
plt.xlabel("Type of restaurant")
plt.show()
```

```python
# Conclusion: According to survey, Majority type of restaurants order from Dining
```

# How many votes has each type of restaurant received from customers?

```python
data_merged = df1.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"Votes":data_merged})
plt.plot(result,c="red",marker="o")
plt.xlabel("Type Of Restaurant",size=15,c="blue")
plt.ylabel("Votes",size=15,c="blue")
plt.grid()
plt.show()
```

```python
# Conclusion: According to survey, Dining restaurants has the maximum votes
```

# What are the ratings that the majority of restaurants have received?

```python
result = df1["name"].value_counts()
```

```python
sns.displot(data=df1,x=df1["rate"])
plt.xlabel("Number of Ratings",size=15)
plt.ylabel("Countings",size=15)
plt.show()
```

```python
# Conclusion: Majority of restaurants have received a number of ratings in between 3.50 and 4.00 
```

# Zomato has observed that most couples order most of their food online. What is their average spending on each order?

```python
sns.countplot(data=df1,x=df1["approx_cost(for two people)"])
plt.xlabel("Average Spending", size=15)
plt.ylabel("Countings", size=15)
plt.show()
```

```python
# Conclusion: Average spending on each order is about Rs 300
```

# Which mode(online/offline) has received the maximum ratings

```python
plt.figure(figsize= (6,6))
sns.boxplot(data=df1,x="online_order",y="rate")
plt.show()
```

```python
plt.figure(figsize= (6,6))
sns.boxplot(data=df1,x="book_table",y="rate")
plt.show()
```

```python
# Conclusion: Online mode has received maximum ratings
```

# Which type of restaurants received more offline orders so that Zomato can produce customers with some good offers

```python
pivoted = df1.pivot_table(index="listed_in(type)", columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivoted, annot=True,fmt='d')
plt.title("Heat Map")
plt.xlabel("Online-Order")
plt.show()
```

```python
#Conclusion: According to survey, 'Dining' type of restaurants received more offline orders and 'Cafes' type of restaurants received more online orders 
```
